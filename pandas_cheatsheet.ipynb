# groupby()

## Use when

Dùng khi muốn nhóm dữ liệu theo một cột và tính toán thống kê cho từng nhóm.

## Syntax

```python
df.groupby("column_name").mean()
```

## Code Example

```python
import pandas as pd

employees = pd.DataFrame({
    "Department": ["IT", "IT", "HR", "HR", "Marketing"],
    "Salary": [1000, 1200, 900, 950, 1100]
})

employees.groupby("Department")["Salary"].mean()
```

## Explanation

Đoạn code trên nhóm nhân viên theo phòng ban và tính mức lương trung bình của từng phòng.

# merge()

## Use when

Dùng khi muốn kết hợp dữ liệu từ nhiều DataFrame dựa trên cột chung.

## Syntax

```python
pd.merge(df1, df2, on="column_name")
```

## Code Example

```python
import pandas as pd

students = pd.DataFrame({
    "ID": [1, 2, 3],
    "Name": ["An", "Binh", "Cuong"]
})

scores = pd.DataFrame({
    "ID": [1, 2, 3],
    "Score": [85, 90, 88]
})

pd.merge(students, scores, on="ID")
```

## Explanation

Đoạn code trên kết hợp thông tin sinh viên và điểm số dựa trên cột ID.

# pivot_table()

## Use when

Dùng để tạo bảng tổng hợp dữ liệu và tính toán thống kê nhanh.

## Syntax

```python
pd.pivot_table(df, values="value", index="column")
```

## Code Example

```python
import pandas as pd

sales = pd.DataFrame({
    "Department": ["IT", "IT", "HR", "HR"],
    "Sales": [500, 700, 300, 400]
})

pd.pivot_table(sales, values="Sales", index="Department", aggfunc="mean")
```

## Explanation

Đoạn code trên tạo bảng tổng hợp doanh thu trung bình của từng phòng ban.

# apply()

## Use when

Dùng để áp dụng một hàm cho từng giá trị trong cột hoặc hàng.

## Syntax

```python
df["column"].apply(function)
```

## Code Example

```python
import pandas as pd

employees = pd.DataFrame({
    "Name": ["An", "Binh", "Cuong"],
    "Salary": [1000, 1200, 900]
})

employees["Bonus"] = employees["Salary"].apply(lambda x: x * 0.1)

employees
```

## Explanation

Đoạn code trên tính tiền thưởng bằng 10% lương cho từng nhân viên.

# melt()

## Use when

Dùng để chuyển dữ liệu từ dạng rộng (wide format) sang dạng dài (long format).

## Syntax

```python
pd.melt(df, id_vars=["column"])
```

## Code Example

```python
import pandas as pd

scores = pd.DataFrame({
    "Name": ["An", "Binh"],
    "Math": [8, 9],
    "English": [7, 8]
})

pd.melt(scores, id_vars=["Name"], var_name="Subject", value_name="Score")
```

## Explanation

Đoạn code trên chuyển bảng điểm từ dạng nhiều cột môn học sang dạng danh sách môn học.

# sort_values()

## Use when

Dùng để sắp xếp dữ liệu theo giá trị của một cột.

## Syntax

```python
df.sort_values(by="column")
```

## Code Example

```python
import pandas as pd

products = pd.DataFrame({
    "Product": ["Laptop", "Mouse", "Keyboard"],
    "Price": [1500, 20, 50]
})

products.sort_values(by="Price", ascending=False)
```

## Explanation

Đoạn code trên sắp xếp sản phẩm theo giá giảm dần.

# dropna()

## Use when

Dùng để xóa các dòng chứa giá trị thiếu (NaN).

# fillna()

## Use when

Dùng để thay thế giá trị thiếu bằng giá trị khác.

## Explanation
